#Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x =[1, 2, 3, 5]
y= [5, 6,7,8]
plt.plot(x, y)
plt.show()
```

```python
x =[1, 2, 3, 5]
y= [5, 6,7,8]
c = "r"
plt.plot(x, y, c)
plt.show()
```

```python
x = [5, 23, 67, 86, 43]
y = [23, 55, 8, 34, 65]
plt.plot(x, y, marker = "o")
plt.show()
```

```python
x = [12,13,8,5,3]
y = [19,12,15,12,42]
plt.plot(x,y)
plt.plot(y, marker = ">")
plt.plot(x, marker = "o")
plt.show()
```
