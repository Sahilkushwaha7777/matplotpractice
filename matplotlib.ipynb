#Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
x =[1, 2, 3, 5]
y= [5, 6,7,8]
plt.plot(x, y)
plt.show()
```

```python
x =[1, 2, 3, 5]
y= [5, 6,7,8]
c = "r"
plt.plot(x, y, c)
plt.show()
```

```python
x = [5, 23, 67, 86, 43]
y = [23, 55, 8, 34, 65]
plt.plot(x, y, marker = "o")
plt.show()
```

```python
x = [12,13,8,5,3]
y = [19,12,15,12,42]
plt.plot(x,y)
plt.plot(y, marker = ">")
plt.plot(x, marker = "o")
plt.show()
```

# bar graph

```python
x =[1, 2, 3, 4]
y= [5, 6,7,8]

plt.bar(x,y)
plt.show()
```

```python
x =[1, 2, 3, 4]
y= [5, 6,7,8]
c=["r","g","b","y"]
plt.bar(x,y,color = c)
plt.show()
```

```python
import numpy as np
x= np.array(["A",'B','C','D'])
Y = np.array([1,3,4,5])
plt.barh(x,y)
plt.show()
```

```python
x= np.array(["A",'B','C','D'])
Y = np.array([1,3,4,5])
plt.barh(x,y, color = "red")
plt.show()
```

```python
x= np.array(["A",'B','C','D'])
Y = np.array([1,3,4,5])
plt.bar(x,y, color = "pink", width = 0.2)
plt.show()
```

```python
x= np.array(["A",'B','C','D'])
Y = np.array([1,3,4,5])
plt.barh(x,y, color = "green", height = 0.5)
plt.show()
```

#scatter plot

```python
x = [1,2,3,4,5,6,7]
y = [4,2,3,5,6,7,8]
plt.title("scatter plot")
plt.xlabel("math")
plt.ylabel("science")
plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [4,2,3,5,6,7,8]
plt.title("red color")
plt.xlabel("number")
plt.ylabel("month")
c = ["red", "blue", "green", "yellow", "pink", "orange", "magenta"]
plt.scatter(x,y, color = c)
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [4,2,3,5,6,7,8]
plt.title("red color")
plt.xlabel("number")
plt.ylabel("month")
plt.scatter(x,y, color = "r")
plt.show()
```

```python
x = [1,2,3,4,5,6,7]
y = [4,2,3,5,6,7,8]
plt.title("red color")
plt.xlabel("number")
plt.ylabel("month")
c = ["red", "blue", "green", "yellow", "pink", "orange", "magenta"]
plt.scatter(x,y, color = c, s= 1000)
plt.show()
```
